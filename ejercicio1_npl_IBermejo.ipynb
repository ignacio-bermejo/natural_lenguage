```python
#Creo un diccionario de palabras unicas y veo cuantas veces se repiten en el texto
#PERO ME SALEN LAS PUTAS FRASES Y NO POR PALABRAS!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
numero_palabras = dict.fromkeys(palabras_unicas, 0)
for palabra in dataset[0]:
    numero_palabras[palabra] += 1

print(numero_palabras)
```

```python
#Asi hago la formula matematica de TF-IDF, en la que saca el numero de veces en el que la palabra aparece en el texto dividido entre el num total
#de palabras en el texto

def computeTF(diccionario_palabra, Bag_Of_Words):
    tf_dict = {}
    Contar_BagOfWords = len(Bag_Of_Words)
    for palabra, contar in diccionario_palabra.items():
        tf_dict[palabra] = contar / float(Contar_BagOfWords)
    return tf_dict
```

```python
#APARECE LA PROBABILIDAD ASOCIADA A LA FRASE, HABLARLO CON ALGUIEN
tf_texto = computeTF(numero_palabras, dataset[0])
print(tf_texto)
```

```python
#Hgo la frecuencia inversa, IDF(Inverse Data Frequency), esto es el numero logaritmico del documento dividido entre el numero de documentos que conteien
#la palabra W.
#IDF determiena el peso de las palabras raras a lo largo del cuerpò del documento

import math

def computeIDF(documentos):
    n = len(documentos)

    idf_diccionario = dict.fromkeys(documentos[0].keys(),0)
    for documento in documentos:
        for palabra, val in documento.items():
            if val > 0:
                idf_diccionario[palabra] += 1

    for palabra, val in idf_diccionario.items():
        idf_diccionario[palabra] = math.log(n / float(val))
    return idf_diccionario
```

```python
idf = computeIDF([numero_palabras])
```

```python
#Ahora junto TF y IDF  multiplicandolos para crear TF-IDF

def computeTFIDF(tfBagOfWords, idf):
    tfidf = {}
    for palabra, val in tfBagOfWords.items():
        tfidf[palabra] = val * idf[palabra]
    return tfidf
```

```python
tfidf = computeTFIDF(tf_texto, idf)
df = pd.DataFrame([tfidf])
```

```python
#Tokenizo las palabras de la columna Texto

texto = dataset.texto

tokenizer = text.Tokenizer()
tokenizer.fit_on_texts(texto)
```

```python
w2id = tokenizer.word_index
id2w = {v:k for k, v in w2id.items()}

vocab_size = len(w2id) + 1 
embed_size = 100
```

```python
#Asigno un numero unico a cada palabra de cada frase

wids = [[w2id[w] for w in text.text_to_word_sequence(doc)] for doc in texto]
print('Vocabulary Size:', vocab_size)
print('Vocabulary Sample:', list(w2id.items())[:10])
```

```python
#Generate skip grams
skip_grams = [skipgrams(wid, vocabulary_size=vocab_size, window_size=10) for wid in wids]

#Veo los skip grams
pairs, labels = skip_grams[0][0], skip_grams[0][1]
for i in range(10):
    print("({:s} ({:d}), {:s} ({:d})) -> {:d}".format(
          id2w[pairs[i][0]], pairs[i][0], 
          id2w[pairs[i][1]], pairs[i][1], 
          labels[i]))
```

```python
stop = stopwords.words('english')
```

```python
#Construyo el modelo
dataset['texto_limpio'] = dataset['texto'].apply(
    lambda x: ' '.join(
        [word.lower() for word in x.split() 
        if word not in (stop) and word.isalnum()]
        )
    )
dataset.head(4)

lb = LabelEncoder()
sentimientos = lb.fit_transform(dataset.sentimiento)
aleat = np.random.choice(dataset.shape[0], 2000, replace=False)
X = dataset.texto_limpio.iloc[aleat]
X_test = dataset.texto_limpio.iloc[~aleat]
y = sentimientos[aleat]
y_test = sentimientos[~aleat]
```

```python
input_a = Input(shape=1)
embedding_a = Embedding(
    input_dim = vocab_size,
    output_dim = embed_size,
    input_length=1)(input_a)

flatten_a = Flatten()(embedding_a)

input_b = Input(shape=1)
embedding_b = Embedding(
    input_dim = vocab_size,
    output_dim = embed_size,
    input_length=1)(input_b)
flatten_b = Flatten()(embedding_b)

prodes = dot([flatten_a, flatten_b], axes=1)

densa = Dense(1,activation="sigmoid")(prodes)

modelo = Model(inputs=[input_a, input_b], outputs=[densa])
modelo.compile(loss="binary_crossentropy", metrics=["accuracy"])
modelo.summary()
```
